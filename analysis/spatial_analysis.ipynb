```python
import pandas as pd
import folium
```

```python
flights_weather_information_df = pd.read_csv('../data/intermediate/cleaning/flights_spatial_subset.csv')
```

```python
# Create a map centered at the origin coordinates
m = folium.Map(location=[flights_weather_information_df['ORIGIN_LATITUDE'][0], flights_weather_information_df['ORIGIN_LONGITUDE'][0]], zoom_start=10)

for index, flight in flights_weather_information_df.iterrows():
    origin = (flight['ORIGIN_LATITUDE'], flight['ORIGIN_LONGITUDE'])
    point_50km = (flight['LATITUDE_50KM'], flight['LONGITUDE_50KM'])
    point_100km = (flight['LATITUDE_100KM'], flight['LONGITUDE_100KM'])
    destination = (flight['DESTINATION_LATITUDE'], flight['DESTINATION_LONGITUDE'])

    # Calculate flight path coordinates
    flight_path = [origin, destination]

    # Add flight path to the map
    folium.PolyLine(locations=flight_path, color='blue', weight=1, opacity=0.3).add_to(m)

    # Add markers to the map
    folium.CircleMarker(location=point_50km, radius=5, color='red', fill=True, fill_color='red').add_to(m)
    folium.CircleMarker(location=point_100km, radius=5, color='green', fill=True, fill_color='green').add_to(m)
```
